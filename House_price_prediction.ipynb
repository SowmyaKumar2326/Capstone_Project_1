### KNN_Regression_Model

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_squared_error
```

```python
# Splitting the data into Train and Test
X = House_data[['bath','balcony','total_sqft_int','bhk','price_per_sqft','area_typeSuper built-up  Area','area_typeBuilt-up  Area',
       'area_typePlot  Area','availability_Ready To Move']]
y = House_data['price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1234)
```

```python
X_train.shape,y_train.shape
```

```python
y_test.shape,y_test.shape
```

### Fitting the Model

```python
model = KNeighborsRegressor()
model_fit = model.fit(X_train,y_train)
pred = model_fit.predict(X_test)
pred
```

```python
result = r2_score(y_test,pred)
result
```

```python
Accu = round(result * 100,2)
```

```python
print(f'The Accuracy of the model is {Accu}%.')
```

```python
rmse = mean_squared_error(y_test,pred)
print('The RMSE value is:',rmse)
```

```python

error = []
k = []
for i in range(1,10,2): 
    model = KNeighborsRegressor(n_neighbors=i)
    model_fit = model.fit(X_train,y_train)
    err  = 1 - round(model_fit.score(X_train,y_train),2)
    error.append(err)
    k.append(i)
```

```python
pd.DataFrame({'K':k,'error':error})
```

### Saving the model

```python
import joblib
joblib.dump(model,'House_price_prediction.sav')
```
